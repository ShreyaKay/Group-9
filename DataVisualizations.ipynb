```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv(r"C:\Users\fayee\Downloads\software_engineers_gender_race_education_salary.csv")

# Strip spaces from column names (if needed)
df.columns = df.columns.str.strip()

# Create the figure and axis
plt.figure(figsize=(12, 12))  # Increased figure size to make it more readable

# Boxplot for Gender and Education Level (Top plot)
plt.subplot(2, 1, 1)  # 2 rows, 1 column, first subplot
sns.boxplot(data=df, x='Education', y='totalyearlycompensation', hue='gender', palette='Set2')
plt.title("Impact of Gender and Education on Compensation", fontsize=14)
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Total Yearly Compensation (USD)", fontsize=12)
plt.xticks(fontsize=10)  # No rotation for x-axis labels
plt.legend(title='Gender', fontsize=10)

# Boxplot for Race and Education Level (Bottom plot)
plt.subplot(2, 1, 2)  # 2 rows, 1 column, second subplot
sns.boxplot(data=df, x='Education', y='totalyearlycompensation', hue='Race', palette='Set1')
plt.title("Impact of Race and Education on Compensation", fontsize=14)
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Total Yearly Compensation (USD)", fontsize=12)
plt.xticks(fontsize=10)  # No rotation for x-axis labels
plt.legend(title='Race', fontsize=10)

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Calculate the average compensation by Race, Gender, and Education Level
race_gender_edu_avg = df.groupby(['Race', 'gender', 'Education'])['totalyearlycompensation'].mean().reset_index()

# Create the plot with separate bars for Education level and Gender, Race on x-axis
plt.figure(figsize=(16, 8))

# Create the plot with separate bars for Education level, Gender, and Race on x-axis
ax = sns.barplot(data=race_gender_edu_avg, x='Race', y='totalyearlycompensation', hue='gender', dodge=True, palette='Set2', errorbar=None)

# Overlay the education level on the same plot with another hue (Education level)
sns.barplot(data=race_gender_edu_avg, x='Race', y='totalyearlycompensation', hue='Education', dodge=True, palette='viridis', errorbar=None)

# Add numerical values on top of each bar
for p in ax.patches:
    height = p.get_height()  # Get the height of the bar (compensation value)
    ax.text(p.get_x() + p.get_width() / 2, height + 1000, f'{height:.0f}',  # Adjust the height to place the value above the bar
            ha='center', va='bottom', fontsize=10, color='black')

# Title and labels
plt.title("Average Compensation by Race, Gender, and Education Level", fontsize=14)
plt.xlabel("Race", fontsize=12)
plt.ylabel("Average Total Yearly Compensation (USD)", fontsize=12)
plt.xticks(rotation=0, fontsize=10)

# Show the plot
plt.legend(title='Gender and Education Level', fontsize=10)
plt.tight_layout()
plt.show()
```

```python

# Calculate the average compensation by Race, Gender, and Education Level
race_gender_edu_avg = df.groupby(['Race', 'gender', 'Education'])['totalyearlycompensation'].mean().reset_index()

# Calculate the mean and median for each race group
mean_values = df.groupby('Race')['totalyearlycompensation'].mean()
median_values = df.groupby('Race')['totalyearlycompensation'].median()

# Create the plot with separate bars for Education level, Race on x-axis, and Gender as hue
plt.figure(figsize=(14, 8))
ax = sns.barplot(data=race_gender_edu_avg, x='Race', y='totalyearlycompensation', hue='Education', palette='viridis', errorbar=None)

# Add numerical values on top of each bar
for p in ax.patches:
    height = p.get_height()  # Get the height of the bar (compensation value)
    ax.text(p.get_x() + p.get_width() / 2, height + 1000, f'{height:.0f}',  # Adjust the height to place the value above the bar
            ha='center', va='bottom', fontsize=10, color='black')

# Add numerical annotations for mean and median values around 100,000 on the y-axis
for i, label in enumerate(ax.get_xticklabels()):
    x_pos = label.get_position()[0]  # Get the x-position of each race label
    
    # Set the position for the mean and median values around 100,000 on the y-axis
    mean_value = mean_values[ax.get_xticklabels()[i].get_text()]
    median_value = median_values[ax.get_xticklabels()[i].get_text()]

    # Positioning the mean just below the median, 1 cm apart
    ax.text(x_pos, 50000, f'Mean: {mean_value:,.0f}', ha='center', va='bottom', fontsize=10, color='black')
    ax.text(x_pos, 35000, f'Median: {median_value:,.0f}', ha='center', va='bottom', fontsize=10, color='black')

# Title and labels
plt.title("Average Compensation by Race, Gender, and Education Level", fontsize=14)
plt.xlabel("Race", fontsize=12)
plt.ylabel("Average Total Yearly Compensation (USD)", fontsize=12)

# Show the plot
plt.legend(title='Education Level', fontsize=10)
plt.tight_layout()
plt.show()
```
